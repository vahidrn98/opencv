```python
import numpy as np
import cv2
```

```python
img = cv2.imread('index.jpg',cv2.IMREAD_COLOR)
cv2.line(img,(0,0),(225,225),(255,0,0),3)# image-starting point-end point for image size-color(bgr)-border
```

```python
cv2.rectangle(img,(15,25),(35,55),(0,255,0),1) #drawing a rectangle
```

```python
cv2.circle(img,(110,110),50,(0,255,255),1) # drawing a circle
cv2.circle(img,(200,200),50,(0,255,255),-1) #filled circle
```

```python
# in order to create a polygon we have to delacre some points to get connected forming the shape
pts = np.array([[0,20],[30,37],[93,50],[23,60],[70,65]],np.int32)
pts = pts.reshape((-1,1,2))
cv2.polylines(img,[pts],True,(255,255,255),1)
```

```python
#writing text
font=cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'opencv tutorial',(0,200),font,1,(255,255,255),2,cv2.LINE_AA) # 1:size,2:thickness
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)#keyboard interupt
cv2.destroyAllWindows()
```
